## Step 1: Create a data frame that contains the total quantity of each product purchased by each customer.

You will need to group by CustomerID and ProductName and then sum the Quantity field.

```python
import numpy as np
import pandas as pd
```

```python
df2 = pd.DataFrame(data.groupby(["CustomerID","ProductName"]).Quantity.sum()).reset_index() #creo que este es
```

## Step 2: Use the `pivot_table` method to create a product by customer matrix.

The rows of the matrix should represent the products, the columns should represent the customers, and the values should be the quantities of each product purchased by each customer. You will also need to replace nulls with zeros, which you can do using the `fillna` method.

```python
#Ejemplo de la documentación
#table = pd.pivot_table(df, values='D', index=['A', 'B'],
#                     columns=['C'], aggfunc=np.sum, fill_value=0)

pivot_table = pd.pivot_table(df2, values="Quantity", index=["ProductName"],
                            columns=["CustomerID"], fill_value=0)  
```

```python
pivot_table
```

## Step 3: Create a customer similarity matrix using `squareform` and `pdist`. For the distance metric, choose "euclidean."

```python
from scipy.spatial.distance import pdist, squareform
```

```python
#Primer intento
customer_similarity_m = squareform(pdist(pivot_table, 'euclidean'))
customer_similarity_m
```

```python
customer_similarity_m = pd.DataFrame(customer_similarity_m) #Lo hacemos como df para que se vea bonito
```

```python
customer_similarity_m
```

```python
customer2 = pd.DataFrame(1/(1+ customer_similarity_m)) #Lo dividimos entre (1/x)
customer2

#Esta matriz no tiene el número de cliente. DIABLOS
```

```python
#Segundo intento
cust_similar = pd.DataFrame(1/(1 + squareform(pdist(pivot_table.T, 'euclidean'))), index=pivot_table.columns, columns=pivot_table.columns)
cust_similar.head(3)
```

## Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.

```python
#Voy a escoger al customer 200

usuarios_parecidos = pd.DataFrame(cust_similar.loc[200].sort_values(ascending=False)).head(5).reset_index()
```
